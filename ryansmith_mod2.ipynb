Chapter 3.10.3 - Reading requirement statements and pseudocode: Building Python script from it

```python
"""Class avereage program with sequence-controlled repetition"""

#initialization phase
total = 0 
grade_counter = 0
grades = [98, 76, 71, 87, 83, 90, 57, 79, 82, 94]  #set counters to 0 and list of 10 grades

#processing phase:

for grade in grades:
    total += grade  # this adds the grade to the current total
    grade_counter += 1  #this will increase one as each grade is processed
    
#termination phase
average = total / grade_counter
print(f'Class average is {average}')
```

```python
#Practicing with sentinal-controlled iterations - Chapter 3.11

total = 0
grade_counter = 0

grade = int(input('Enter grade, or enter -1 to end: '))

while grade != -1:
    total += grade
    grade_counter += 1
    grade = int(input('Enter grade, or enter -1 to end: '))
    
if grade_counter != 0:
    average = total / grade_counter
    print (f'Class average is {average:.2f}')
else:
    print('No grades were entered')
```

6: Calculating Measures of Central Tendancy

```python
# Using data from pg 110-111
#Performing Count, Sum, and Mean

grades = [85, 93, 45, 89, 85]

print('The sum of the grades is: ', sum(grades))
print('The count of the grades is: ', len(grades))

print('The mean of the grades is: ', sum(grades) / len(grades))
```

7: Calculating Measures of Central Tendancies with Statistics Module

```python
#Using the same data as above to use the statistics module to calculate
#Performing calculations to get Mean, Median, and Mode

import statistics

grades = [85, 93, 45, 89, 85]

print('The mean of the grades is: ', statistics.mean(grades))
print('The median of the grades is: ', statistics.median(grades))
print('The mode of the grades is: ', statistics.mode(grades))
```
